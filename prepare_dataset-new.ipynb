```python
import cv2
import os
import matplotlib.pyplot as plt

# Define the directories for images and labels
image_dir = 'dataset/images'
label_dir = 'dataset/labels'

f = open("license_plate.csv", "w")

# Get a list of image files in the image directory
image_files = os.listdir(image_dir)

for image_file in image_files:
    # Load the original image
    image_path = os.path.join(image_dir, image_file)
    image = cv2.imread(image_path)
    # plt.imshow(image)

    # Construct the label file path
    label_file = image_file.replace('.jpg', '.txt')
    label_path = os.path.join(label_dir, label_file)

    # Read the label annotation from the file
    with open(label_path, 'r') as file:
        line = file.readline().strip()
        label_parts = line.split()

    if len(label_parts) < 5:
        print(f"Skipping {image_file} due to invalid label data.")
        continue

    x, y, width, height = map(float, label_parts[1:])

    image_height, image_width, _ = image.shape
    x_pixel = int(x * image_width)
    y_pixel = int(y * image_height)
    width_pixel = int(width * image_width)
    height_pixel = int(height * image_height)

    # Calculate the top-left and bottom-right coordinates
    x1 = x_pixel - (width_pixel // 2)
    y1 = y_pixel - (height_pixel // 2)
    x2 = x_pixel + (width_pixel // 2)
    y2 = y_pixel + (height_pixel // 2)

    # print("{0},{1},{2},{3},{4}".format(image_file,x, y, width, height))
    f.write("{0},{1},{2},{3},{4}\n".format(image_file,x1, x2, y1, y2))

print("Cropping complete. Cropped images saved in the 'cropped_images' directory.")
f.close()
```
